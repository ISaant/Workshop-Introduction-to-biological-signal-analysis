## Libraries 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Fs = 80 #Sampling Frequency way over the minumun necessary
f = [1,3,5] #Frequencies in the signal
t=np.arange(0,2,1/Fs) #Samples array (# of datapoints in two seconds)
y=np.zeros(len(t))
for freq in f:
    x=np.sin(2 * np.pi * freq * t)
    y = y+x
    plt.plot(t, x,'o--',alpha=.3)
plt.plot(t, y,'o-')
plt.xlabel('sample(n)')
plt.ylabel('voltage(V)')
plt.show()
```

```python
Fs = 10 # Minimum Sampling Frequency necessary 
f = [1,3,5] #Frequencies in the signal
t=np.arange(0,2,1/Fs) #Samples array (# of datapoints in two seconds)
y=np.zeros(len(t))
for freq in f:
    x=np.sin(2 * np.pi * freq * t)
    y = y+x
    plt.plot(t, x,'o--',alpha=.3)
plt.plot(t, y,'o-')
plt.xlabel('sample(n)')
plt.ylabel('voltage(V)')
plt.show()
```

```python
Fs = 8 # Below the Sampling Frequency necessary according to the Nyquist Theorem.  
f = [1,3,5] #Frequencies in the signal
t=np.arange(0,2,1/Fs) #Samples array (# of datapoints in two seconds)
y=np.zeros(len(t))
for freq in f:
    x=np.sin(2 * np.pi * freq * t)
    y = y+x
    plt.plot(t, x,'o--',alpha=.3)
plt.plot(t, y,'o-')
plt.xlabel('sample(n)')
plt.ylabel('voltage(V)')
plt.show()
```
